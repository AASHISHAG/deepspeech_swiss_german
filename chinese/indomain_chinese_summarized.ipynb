```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
df_new = pd.DataFrame(columns = ['aidatatang', 'aishell','magicdata','prime','stcmds','thchs30'])
```

```python
with open('in_aidatatang.json', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
    
df_new['Aidatatang']= df['res'].str.len()/df['src'].str.len()
```

```python
with open('in_aishell.json', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
    
df_new['Aishell']= df['res'].str.len()/df['src'].str.len()
```

```python
with open('in_magicdata.json', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
    
df_new['Magicdata']= df['res'].str.len()/df['src'].str.len()
```

```python
with open('in_prime.json', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
    
df_new['Prime']= df['res'].str.len()/df['src'].str.len()
```

```python
with open('in_stcmds.json', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
    
df_new['Stcmds']= df['res'].str.len()/df['src'].str.len()
```

```python
with open('in_thchs30.json', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
    
df_new['Thchs30']= df['res'].str.len()/df['src'].str.len()
```

```python
ax = sns.kdeplot(df_new['len_ratio_in_aidatatang'], shade=True)
sns.kdeplot(df_new['len_ratio_in_aidatatang'], shade=True)
sns.kdeplot(df_new['len_ratio_in_magicdata'], shade=True)
sns.kdeplot(df_new['len_ratio_in_prime'], shade=True)
sns.kdeplot(df_new['len_ratio_in_stcmds'], shade=True)
sns.kdeplot(df_new['len_ratio_in_thchs30'], shade=True)

#for axis in ['top','bottom','left','right']:
#  ax.spines[axis].set_linewidth(0.1)
#ax.axes([0, 0.6, 1, 1])
sns.set_style("whitegrid")
ax.set(xlabel='Audio Length (seconds)', ylabel='Density')
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
def replace1(x):
    if x >= 1.2:
        return 1.2
    elif x <= 0.5:
        return 0.5
    else: 
        return x
```

```python
df_new['Aidatatang'] = df_new['Aidatatang'].apply(replace1)
df_new['Magicdata'] = df_new['Magicdata'].apply(replace1)
df_new['Prime'] = df_new['Prime'].apply(replace1)
df_new['Stcmds'] = df_new['Stcmds'].apply(replace1)
df_new['Thchs30'] = df_new['Thchs30'].apply(replace1)
#df_new['Aishell'] = df_new['len_ratio_in_aishell'].apply(replace1)
```
