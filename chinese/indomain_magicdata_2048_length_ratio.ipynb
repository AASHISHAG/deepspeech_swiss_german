```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
```

```python
with open('in_magic_2048.json', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
    
df['len_ratio']= df['res'].str.len()/df['src'].str.len()
```

```python
df.columns
```

```python
#df[['wav_filename','src', 'res', 'wer', 'word_distance', 'word_length', 'cer', 'char_distance', 'char_length', 'len_ratio']]
```

```python
ax = sns.distplot(df['cer'], bins=10) # blue
sns.distplot(df['cer'], bins=10) # orange
ax.set(xlabel='Error Rate', ylabel='Density')
plt.show()
```

```python
df_no_issue = df[df['len_ratio'] > 0.75]
print(len(df))
print(len(df_no_issue))
ax = sns.distplot(df_no_issue['cer'], bins=10) # blue
sns.distplot(df_no_issue['cer'], bins=10) # orange
ax.set(xlabel='Error Rate', ylabel='Density')
plt.show()
```

```python
print(df['cer'].mean())
print(df_no_issue['cer'].mean())
print()
print(df['wer'].mean())
print(df_no_issue['wer'].mean())
```
