```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
```

```python
with open('C:/Users/Aashish Agarwal/Desktop/out_swiss.txt', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
```

```python
df[['wav_filename','src', 'res', 'wer', 'word_distance', 'word_length', 'cer', 'char_distance', 'char_length']]
```
