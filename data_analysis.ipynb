```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
```

```python
with open('out_swiss.txt', 'r', encoding='utf-8') as file_read:
    data = file_read.read()
    loaded_json = json.loads(data)
    df=pd.DataFrame.from_dict(loaded_json)
    
df['len_ratio']= df['src'].str.len()/df['res'].str.len()
```

```python
df.columns
```

```python
df[['wav_filename','src', 'res', 'wer', 'word_distance', 'word_length', 'cer', 'char_distance', 'char_length']]
```

```python
ax = sns.distplot(df['wer'], bins=10) # blue
sns.distplot(df['cer'], bins=10) # orange
ax.set(xlabel='Error Rate', ylabel='Density')
plt.show()
```

```python
df_no_issue = df[df['len_ratio'] < 1.3]
print(len(df))
print(len(df_no_issue))
ax = sns.distplot(df_no_issue['wer'], bins=10) # blue
sns.distplot(df_no_issue['cer'], bins=10) # orange
ax.set(xlabel='Error Rate', ylabel='Density')
plt.show()
```

```python
print(df['cer'].mean())
print(df_no_issue['cer'].mean())
print()
print(df['wer'].mean())
print(df_no_issue['wer'].mean())
```

```python
sns.pairplot(df[['wer', 'word_distance', 'word_length', 'cer', 'char_distance', 'char_length']])
```

```python
myBasicCorr = df.corr()
sns.heatmap(myBasicCorr, annot = True)
fig = plt.gcf()  # or by other means, like plt.subplots
figsize = fig.get_size_inches()
sns.set(font_scale=3)
fig.set_size_inches(figsize * 2)  # scale current size by 1.5
```
